```python
import requests
import json
from bs4 import BeautifulSoup
from csv import writer
url="https://yallacompare.com/ajax/moreProductsData/?country=uae&category=credit-cards&lang=en&salary=&features=&bank=&sortByColumn=&sortByOrder=&pageNumber=2&itemsPerPage=10&isRamadan=false&loanAmount=0&term=0&downPayment=0"
data=requests.get(url)
credit=json.loads(data.content)
```

```python
credit
```

```python
credit['result']['products']
```

```python
date_header=['title','FlatRate','MinimunSalary','MaximumTerm','card_image','feature','feature_slug','isSalaryTransfer',
            'Card name','bank_name','bank_logo','slug',]
credit_card_data=[]
for x in credit['result']['products']:
    featurs=[]
    feature_slug=[]
    for y in x['translation']['features']:
        featurs.append(y['name'])
        feature_slug.append(y['slug'])
    card_offer=f"https://yallacompare.com/ajax/moreDetails/?category=credit-cards&slug={x['translation']['slug']}&lang=en&country=uae&v2=true"
    credit_card_data.append([x['translation']['title'],x['flatRate'],x['minimumSalary']
                         ,x['maximumTerm'],
                      ,f"https:{x['translation']['image']}","|".join(featurs), "|".join(feature_slug),                            
                        x['translation']['isSalaryTransfer'],x['translation']['metaTitle'],
                         x['translation']['client']['name'],f"https{x['translation']['client']['logo']}",
                        x['translation']['client']['slug'],x['translation']['slug'],x['annualFee'],x['detail']['country'],
                         str(card_offer)    
                        ])
```
