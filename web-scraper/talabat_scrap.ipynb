```python
import requests
from bs4 import BeautifulSoup
import json 
from csv import writer
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:89.0) Gecko/20100101 Firefox/89.0'}

error_rest=[]
all_restaurant_data=[]
rest_file=open("latest_resturent_talabat12.csv",'a')
writer_object = writer(rest_file)
writer_object.writerow(['Restaurants id','restaurant name','cuisines','slug','rating','rated by people','review_count','country',"image","Restaurants link","description"])
rest_file.close()
def get_talabat_data(initial,final):
    print(initial,type(initial))
    print(final,type(final))
    rest_file=open("latest_resturent_talabat12.csv",'a')
    writer_object = writer(rest_file)
    URL = f"https://www.talabat.com/uae/restaurants"
    page=requests.get(URL,headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")
    soup_body=soup.find("body")
    rest_data=soup_body.find("script",id="__NEXT_DATA__")
    dta=json.loads(str(rest_data).replace('<script id="__NEXT_DATA__" type="application/json">',"").replace('</script>',""))
    restaurants_data=dta['props']['pageProps']['restaurants']
    for data in restaurants_data[initial:final]:
        save_list=[]
        resturent_slug_url=f"https://www.talabat.com/uae/{data['slug']}"
        resturent_slug_url_page=requests.get(resturent_slug_url,headers=headers)
        if resturent_slug_url_page.status_code == 200:
            print(resturent_slug_url)
            save_list.append(data['id'])
            save_list.append(data['name'])
            save_list.append(data['cuisines'])
            save_list.append(data['slug'])
            find_resturent_rating=BeautifulSoup(resturent_slug_url_page.content, "html.parser")
            try:
                resturent_rating=find_resturent_rating.find("div",class_="rating-number").text
            except:
                resturent_rating=0
            try:    
                resturnet_rating_by_people=find_resturent_rating.find("div",class_="ratings-count").text.split(" ")[1]
            except:
                resturnet_rating_by_people=0
            try:    
                resturnet_rating_by_people_review=find_resturent_rating.find("span",class_="f-500").text.split(" ")[0]
            except:
                resturnet_rating_by_people_review=0
            try:
                resturnet_description=find_resturent_rating.find("div",class_="brand-descriptionstyles__InfoBarContainer-sc-2v5ej3-0")
                descriptions=resturnet_description.find("div",class_="dark-gray")
                des=descriptions.text
            except:    
                des=""    
            save_list.append(resturent_rating)
            save_list.append(resturnet_rating_by_people)
            save_list.append(resturnet_rating_by_people_review)
            save_list.append("UAE")
            save_list.append(data['logo'].split("?")[0])
            save_list.append(resturent_slug_url)
            save_list.append(des)
            all_restaurant_data.append(save_list)
            writer_object.writerow(save_list)
        else:
            print("error")
            error_rest.append(resturent_slug_url)
            print(resturent_slug_url)
            print("error")
    rest_file.close()       
```
